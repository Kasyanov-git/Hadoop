```python
%%file biGRAM.py

import nltk
from nltk.util import ngrams
from nltk.corpus import stopwords
from mrjob.job import MRJob, MRStep
import re

class MRBigram(MRJob):
    def steps(self):
        return [
            MRStep(mapper_init=self.mapper_init,
                   mapper = self.mapper,
                   reducer=self.reducer),
            MRStep(reducer=self.reducer_final)
        ]

    def mapper_init(self):
       nltk.download('punkt')
       nltk.download('stopwords')
    
    def mapper(self, _, line):
       dialogue = line.split('" "')
       text = dialogue[-1].lower()
    
       text = re.sub(r'\W+', ' ', text)
    
       words = nltk.word_tokenize(text)
    
       words = [word for word in words if word not in stopwords.words('english')]
    
       bigrams = list(ngrams(words, 2))
    
       for bigram in bigrams:
           yield (bigram[0] + ' ' + bigram[1]), 1
    
    def reducer(self, bigram, counts):
       yield None,  (bigram, sum(counts))

    def reducer_final(self, _, bigrams):
        top_bigrams = sorted(bigrams, key=lambda x: x[1], reverse=True)[:20]
        for bigram, num in top_bigrams:
            yield bigram, num

if __name__ == '__main__':
   MRBigram.run()
```
